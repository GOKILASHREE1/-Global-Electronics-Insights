```python
import pandas as pd
import mysql.connector

# Load the data from CSV
data = pd.read_csv("cleaned_customers.csv")

# Connect to MySQL server
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="geanalysis"  # Ensure the database name matches
)

# Create a cursor object
mycursor = mydb.cursor()

# Iterate over the DataFrame and insert values into the table
for index, row in data.iterrows():
    sql = '''
    INSERT INTO customers_data (CustomerKey, Gender, Name, City, `State Code`, State, `Zip Code`, Country, Continent, Age)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    '''
    values = (
        row['CustomerKey'],
        row['Gender'],
        row['Name'],
        row['City'],
        row['State Code'],
        row['State'],
        row['Zip Code'],
        row['Country'],
        row['Continent'],
        row['Age']
    )
    
    mycursor.execute(sql, values)

# Commit the transaction
mydb.commit()

# Close the cursor and connection
mycursor.close()
mydb.close()
```

```python
import mysql.connector
import pandas as pd

# Database connection configuration
config = {
    'user': 'root',                  
    'password': '',                  
    'host': 'localhost',
    'database': 'geanalysis'
}

# Establish connection to MySQL
mydb = mysql.connector.connect(**config)
mycursor = mydb.cursor()

# Create the stores_data table
mycursor.execute('''
CREATE TABLE IF NOT EXISTS stores_data (
    StoreKey INT PRIMARY KEY,
    Country VARCHAR(100),
    State VARCHAR(100),
    SquareMeters DECIMAL(10, 2),
    OpenDate DATE
)
''')

# Load the data from CSV
data = pd.read_csv("cleaned_stores.csv")  # Update with the correct path to your CSV file

# Iterate over the DataFrame and insert values into the table
for index, row in data.iterrows():
    sql = '''
    INSERT INTO stores_data (StoreKey, Country, State, SquareMeters, OpenDate)
    VALUES (%s, %s, %s, %s, %s)
    '''
    values = (
        row['StoreKey'],
        row['Country'],
        row['State'],
        row['Square Meters'],
        row['Open Date']
    )
    
    mycursor.execute(sql, values)

# Commit the transaction
mydb.commit()

# Close the cursor and connection
mycursor.close()
mydb.close()
```
